```
#import PUT YOUR CODE HERE
import pandas as pd
import requests
#load data diectly from the url 'https://opendata.ecdc.europa.eu/covid19/casedistribution/csv'
url='https://opendata.ecdc.europa.eu/covid19/casedistribution/csv'

#This is new, till now we used a file, now we give pandas just the link on the internet. 
df = pd.read_csv(url)
```

```
#Use the function sample of pandas with the right value to show 10 random values
#PUT YOUR CODE HERE
df.sample(n=10)
```

```
# Take only useful columns
column_names = ['dateRep', 'cases', 'deaths', 'countriesAndTerritories', 'countryterritoryCode', 'continentExp','popData2019']
df = df[column_names]
# Rename columns to more comfortable names
df = df.rename(columns={
    'dateRep': 'date',
    'countriesAndTerritories': 'country',
    'countryterritoryCode': 'countryCode',
    'continentExp': 'continent',
    'popData2019': 'population'
})
# Convert string to datetime
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y')
#Use the function sample with the value 10 to show 10 random values
#PUT YOUR CODE HERE
df = df.sample(10)
```

```
#Let us look on the data at a specific date: today two years ago
from datetime import datetime
from dateutil.relativedelta import relativedelta
# Get today two years ago as string
today_two_years_ago = (datetime.now()- relativedelta(years=2)).strftime('%Y-%m-%d')
print(today_two_years_ago)
# Get a data frame only for today two years ago
df_today_two_years_ago = df[df.date == today_two_years_ago]
# Preview the first 5 values of df_today using the function head of pandas
df.head(5)
```

```
# Plot a goegraphic illustration. Feed the figure with the data frame of two years ago.
import plotly.express as px
fig = px.scatter_geo(
    df_today_two_years_ago, 
    locations='countryCode',
    color='continent',
    hover_name='country',
    size='cases',
    projection="natural earth",
    title='World COVID-19 Cases for {today}'
)
fig.show()
```

```
# Convert date to string type
df["date"] = df.date.dt.strftime('%Y%m%d')
# Use pandas to sort values by date
df = df.sort_values(by=["date"])
# Some countries does not have code, let's drop all the invalid rows using the function dropna() of pandas
df = df.dropna()
# Preview the data frame
df.head()
```

```
df
```

```
#Let us clean once more
#Take only the cases values which are bigger or equal to zero in your data frame
df = df[df.cases>=0]
```

```
df
```

```
fig = px.scatter_geo(
    df, 
    locations='countryCode',
    color='continent',
    hover_name='country',
    size='cases',
    projection="natural earth",
    title=f'World COVID-19 Cases',
    animation_frame="date"
)
fig.show()
```

```
#Let's make a comulative sum in our data frame
df['cummultive sum'] = df.groupby('country')['cases'].transform(pd.Series.cumsum)
# Plot the last 10 values using the function tail of pandas
df.tail(10)
```

```
# Plot a goegraphic illustration
fig = px.scatter_geo(
    df, 
    locations='countryCode',
    color='continent',
    hover_name='country',
    size='cummultive sum',
    projection="natural earth",
    title=f'World COVID-19 Cases',
    animation_frame="date"
)
fig.show()
```

```
#Let us make a new variable, which measures the number of infections per population.
#We compute the cumulative sum with respect to the population
df['cases_per_pop'] =  df['cummultive sum'] / ( df['population']) 
# I needed to add another line of code because there were some negativ numbers here
df = df[df.cases_per_pop>0]
#Let us look only on Switzerland
df[df.country == "Switzerland"].tail()
```

```
# Plot a goegraphic illustration
fig = px.scatter_geo(
    df, 
    locations='countryCode',
    color='continent',
    hover_name='country',
    size='cases_per_pop',
    projection="natural earth",
    title='World COVID-19 Cases per 1M',
    animation_frame="date"
)
fig.show()
```
